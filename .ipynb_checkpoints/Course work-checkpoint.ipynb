```python
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
from pprint import pprint
```

```python
data = open('data.xml')
md = BeautifulSoup(data) # md - machines data

print(md.findAll("machine"))
# >>> y.foo.bar.type["foobar"]
# u'1'

# >>> y.foo.bar.findAll("type")
# [<type foobar="1"></type>, <type foobar="2"></type>]

# >>> y.foo.bar.findAll("type")[0]["foobar"]
# u'1'
# >>> y.foo.bar.findAll("type")[1]["foobar"]
# u'2'
```

```python
# get r(t) where y - year/number of period; t - age; y-t - year of production

def r(t, y):
    r = []
    try:
        if (y-t) > 0:
            r = md.findAll("machine")[y-t].findAll("summary")[t]["profit"]
        else:
            r = md.findAll("machine")[0].findAll("summary")[t-y]["profit"]
    except IndexError:
        r = '0'
    
    return int(r)
    
r(9,2)
```

```python
# get u(t) where y - year/number of period; t - age; y-t - year of production

def u(t, y):
    u = []
    try:
        u = md.findAll("machine")[y-t].findAll("summary")[t]["servisecost"]
    except IndexError:
        u = '0'
    
    return int(u)
    
u(3,3)
```

```python
# get c(t) where y - year/number of period; t - age; y-t - year of production

def c(t, y):
    c = []
    try:
        c = md.findAll("machine")[y-t].findAll("summary")[t]["delta"]
    except IndexError:
        c = '0'
    
    return int(c)
    
c(3,1)
```

```python
# t - age of the machine, 

def W_tt(year,t,W_next): 
   
    
    save_profit = r(t,year) - u(t,year) + W_next
    change_profit = r(t,year) - u(0,year) - c(t,year) + W_next
    
    
    if save_profit >= change_profit:
        w = save_profit
        x = "save"
    else:
        w = change_profit
        x = "change"
    print(x)
    return w,x

# w = W(9,0,0)
```

```python
def findMax_t():
    arr = md.findAll("machine")

    max_t = len(md.findAll("machine")[0].findAll("summary"))
    for m in arr:
        m1 = m.findAll("summary")
        if max_t < len(m1):
            max_t = len(m1)
            
    return max_t
findMax_t()
```

```python
from prettytable import PrettyTable
def print_table(input):
    #Color
    G = "\033[0;32;40m" # GREEN
    Y = "\033[0;33;40m" # Yellow
    N = "\033[0m" # Reset

    
    t = PrettyTable(['year'] + [i for i in range(t_max+2)])

    row = []
    for k in range(len(input)-1):
        row.append(str(k) + ' year ')
        for w in input[k]:
            w_val = w['w']
            if w['x'] == 'sv':
                val = G + str(w_val) + N
            elif w['x'] == 'ch':
                val = Y + str(w_val) + N
            else:
                val = ''
            
            row.append(val)
        t.add_row(row)
        row = []

    
    print(t)
print_table(w)
```
